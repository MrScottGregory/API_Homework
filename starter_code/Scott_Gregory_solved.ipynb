### Analysis
* In our scatterplot of max city temp vs latitude, we can easily observe that max temps increase as one nears the equator. It is interesting to note that the highest max temps this time of year (May) are found in latitudes just north of the equator.  

* There appears to be less of a correlation between humidity and latitude, as most latitudes include cities with a range of humidity levels.  However, we can note that cities near the equator and approaching the southernmost latitudes seem to have uniformly higher humidity levels at this time.

* Weather factors like cloudiness and wind speed seem to have little correlation with latitude in this data set. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy as ow

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information
base_url = "http://api.openweathermap.org/data/2.5/weather?"

# Build query URL, setting units to imperial to get temps in Fahrenheit
query_url = base_url + "appid=" + api_key + "&units=imperial" + "&q="
#query_url 

# set up lists to hold reponse info
temp = []
humidity = []
cloudiness = []
wind_speed = []
max_temp = []
lat = []
lng = []
date = []
```

```python
# loop through all cities names in city list above 
# identify temp, humidity, cloudiness & wind speed for each city
# append those data points to their respective lists, established above
# print a confirmation line when data points successfully identified
# use exception handling to print notice and move on when city name not identified

for city in cities:
    response = requests.get(query_url + city).json()
    try:
        temp.append(response['main']['temp'])
    except KeyError:
        print("City not found. Skipping...")
    try:
        humidity.append(response['main']['humidity'])
    except KeyError:
        print("City not found. Skipping...")
    try:
        cloudiness.append(response['clouds']['all'])
    except KeyError:
        print("City not found. Skipping...")
    try:
        wind_speed.append(response['wind']['speed'])
    except KeyError:
        print("City not found. Skipping...")
    try:
        max_temp.append(response['main']['temp_max'])
    except KeyError:
        print("City not found. Skipping...")
    try:
        lat.append(response['coord']['lat'])
    except KeyError:
        print("City not found. Skipping...")      
    try:
        lng.append(response['coord']['lon'])
    except KeyError:
        print("City not found. Skipping...")   
    try:
        date.append(response['dt'])
    except KeyError:
        print("City not found. Skipping...")  
        
    print("Data points identified for city of: " + city)
        
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
                
```

```python
# create list of cities that were identified above (not all were found in API search)
# identify list of corresponding country codes for those cities

# create list to hold list of found cities and country names
found_cities = []
countries = []

# create zip to roll up lat and lng coordindates found in loop above
lat_lngs_for_country = zip(lat, lng)

# loop through coordinates, identifying country codes using citipy
# append each country code to the countries list
for coord in lat_lngs_for_country:
    found_city = citipy.nearest_city(coord[0], coord[1]).city_name
    found_cities.append(found_city)
    country = citipy.nearest_city(coord[0], coord[1]).country_code
    countries.append(country)

    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a dictionary with the variables above
weather_dict = {
    "City": found_cities,
    "Cloudiness": cloudiness,
    "Country": countries,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed,
    }

# output into a data frame
raw_weather_data = pd.DataFrame(weather_dict)

# after visualizing, I noticed some Humidity levels were > 100%, which are outliers or errors
# to just look at cities with Humidity less than or equal to 100%:
weather_data = raw_weather_data[raw_weather_data.Humidity <= 100]
weather_data.head()
```

```python
# output to CSV file
weather_data.to_csv("cities.csv", index=False, header=True)
```
